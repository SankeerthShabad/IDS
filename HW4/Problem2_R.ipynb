# Using Sorting Method

```r
#Sorting saturated fat into desending order
Nutrition_sorted <- (Nutrition_subsetDF[order(Nutrition_subsetDF$saturated_fat, decreasing = TRUE),])
Nutrition_sorted
```

### Sorting the Nutrition subsetDF dataframe in decreasing order by saturated fat. Cheesecake has the greatest saturated fat content.

# List of 5 food items

```r
#five food items for highest saturated_fat

head(Nutrition_sorted$food_item)[0:5]
```

### The five foods with the most saturated fat

# Creating new column

```r
#creating saturated_fat_per_gram column in Nutrition_sorted

Nutrition_sorted$saturated_fat_per_gram <- Nutrition_sorted$saturated_fat/Nutrition_sorted$weight_in_grams
Nutrition_sorted
```

### By dividing weight in grams by saturated fat per gram, a new column entitled saturated fat per gram is created.

# Sorting method

```r
#sorting Nutrition_sorted by saturated_fat_per_gram

Nutrition_sorted <-Nutrition_sorted[order(Nutrition_sorted$saturated_fat_per_gram,decreasing=TRUE),]
Nutrition_sorted
```

### Sorting Nutrition subsetDF by saturated fat per gram. From above table we can observe Saturated fat per gram is greatest in butter salted 1TBSP.

# List of 5 food items

```r
#five food items for highest saturated_fat_per_gram

head(Nutrition_sorted$food_item)[0:5]
```

### The five foods with the most saturated fat per gram

# Using summary() method

```r
#summary function

summary(Nutrition_sorted$saturated_fat_per_gram)
```

# Zscore for age


```r
#zscore 

zscore=abs((Nutrition_sorted$saturated_fat_per_gram-mean(Nutrition_sorted$saturated_fat_per_gram))/sd(Nutrition_sorted$saturated_fat_per_gram))
zscore
```

# Detecting Outliers for saturated_fat_per_gram


```r
#boxplot to find outliers

boxplot(Nutrition_sorted$saturated_fat_per_gram,horizontal = TRUE)
```

### Outliers from 0.1 are plainly seen in the graph above.

## Identifying lower and upper limit outliers

```r
#lower and upper bound of outliers

Q1 <- quantile(Nutrition_sorted$saturated_fat_per_gram,0.25)
Q3 <- quantile(Nutrition_sorted$saturated_fat_per_gram,0.75)
IQR <- Q3-Q1
lower_bound <- Nutrition_sorted$saturated_fat_per_gram[Nutrition_sorted$saturated_fat_per_gram<(Q1-(1.5 * IQR))]
upper_bound <- Nutrition_sorted$saturated_fat_per_gram[Nutrition_sorted$saturated_fat_per_gram>(Q3+(1.5 * IQR))]
```

```r
#lower bound for saturated_fat_per_gram

lower_bound
```

```r
#upper bound outlier for saturated_fat_per_gram

upper_bound
```

```r
#maximum upper bound outlier for saturated_fat_per_gram 

max(upper_bound)
```

### To determine the lower and upper bounds of outliers, we can see that there are no lower bound outliers because the value is 0, and the upper bound in saturated fat is 0.507.

```r
#ploting density scaled for saturated_fat_per_gram

par(mfrow=c(1,1))
d <- density(Nutrition_sorted$saturated_fat_per_gram) 
plot(d, main="Kernel Density of saturated_fat_per_gram") 
polygon(d, col="white", border="blue") 
```

### Outliers ranging from 0.2 to 0.5 may be seen in the density figure above due to a slight skew in the data. and the entire range is between 0 and 0.2
